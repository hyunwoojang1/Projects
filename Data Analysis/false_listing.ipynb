## 총 주차 대수 (Total_parking_spots)

```python
def outliner(data, column):
  Q1 = data[column].quantile(0.25)
  Q3 = data[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 -1.5 * IQR
  upper_bound = Q3 +1.5 * IQR
  outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
  return lower_bound, upper_bound, outliers
```

```python
train['total_parking_spots'].describe()
```

```python
sns.barplot(x = 'total_parking_spots', y = 'false_listing_flag', data = train)
plt.xticks(range(0, int(train['total_parking_spots'].max()), 2))
plt.show()
```

```python
lb_TPS, ub_TPS, total_parking_spots_outliers = outliner(train, 'total_parking_spots')
total_parking_spots_outliers
```

```python
lb_TPS, ub_TPS
```

```python
train.loc[ (train['total_parking_spots']> 71.5) | (train['total_parking_spots']< -36.5), 'total_parking_spots'] = 10
```

```python
sns.barplot(x = 'total_parking_spots', y = 'false_listing_flag', data = train, ci = None)
plt.xticks(range(0, int(train['total_parking_spots'].max()), 4))
plt.show()
```
